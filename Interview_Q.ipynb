Q: Order all countries by the year they first participated in the Olympics.
Output the National Olympics Committee (NOC) name along with the desired year.
Sort records in ascending order by year, and alphabetically by NOC.

Company: ESPN

Topics: Aggregations, DataFrame Operations, Filtering & Sorting

```python
import pandas as pd
from google.colab import drive
import warnings # Hide the warnings
warnings.filterwarnings('ignore')
drive.mount('/content/drive')
df=pd.read_csv('/content/drive/My Drive/Colab Notebooks/Interview_Preparation/10184-order-all-countries-by-the-year.csv')
df.head()
```

This code snippet performs the following actions:
1. `df.groupby('noc')['year'].min()`:
   - `df.groupby('noc')`: Groups the DataFrame `df` by the 'noc' (National Olympics Committee) column. This means that all rows belonging to the same NOC are logically grouped together.
   - `['year'].min()`: For each of these NOC groups, it selects the 'year' column and finds the minimum value within that group. This effectively identifies the earliest year each NOC participated in the Olympics.

2. `.reset_index()`:
   - After `groupby()` and `min()`, the 'noc' column becomes the index of the resulting Series. `reset_index()` converts this Series back into a DataFrame, making 'noc' a regular column again alongside the 'year' column.

3. `.sort_values(by=['year', 'noc'])`:
   - This sorts the DataFrame. The primary sorting key is 'year' (ascending by default), meaning countries that first participated earlier will appear first.
   - For countries that first participated in the same year, the secondary sorting key is 'noc' (alphabetically in ascending order), ensuring a consistent order for ties.

4. `print(...)`:
   - Finally, this function prints the resulting DataFrame to the console. The output will show each unique NOC and their earliest participation year, sorted as described above.

```python
print(df.groupby('noc')['year'].min().reset_index().sort_values(by=['year', 'noc']))
```
