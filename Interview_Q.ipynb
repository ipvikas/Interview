Q1: Order all countries by the year they first participated in the Olympics.
Output the National Olympics Committee (NOC) name along with the desired year.
Sort records in ascending order by year, and alphabetically by NOC.

Company: ESPN

Topics: Aggregations, DataFrame Operations, Filtering & Sorting

```python
import pandas as pd
import warnings # Hide the warnings
warnings.filterwarnings('ignore')

df=pd.read_csv('https://raw.githubusercontent.com/ipvikas/Interview/main/10184-order-all-countries-by-the-year.csv')
df.head()
```

This code snippet performs the following actions:
1. `df.groupby('noc')['year'].min()`:
   - `df.groupby('noc')`: Groups the DataFrame `df` by the 'noc' (National Olympics Committee) column. This means that all rows belonging to the same NOC are logically grouped together.
   - `['year'].min()`: For each of these NOC groups, it selects the 'year' column and finds the minimum value within that group. This effectively identifies the earliest year each NOC participated in the Olympics.

2. `.reset_index()`:
   - After `groupby()` and `min()`, the 'noc' column becomes the index of the resulting Series. `reset_index()` converts this Series back into a DataFrame, making 'noc' a regular column again alongside the 'year' column.

3. `.sort_values(by=['year', 'noc'])`:
   - This sorts the DataFrame. The primary sorting key is 'year' (ascending by default), meaning countries that first participated earlier will appear first.
   - For countries that first participated in the same year, the secondary sorting key is 'noc' (alphabetically in ascending order), ensuring a consistent order for ties.

4. `print(...)`:
   - Finally, this function prints the resulting DataFrame to the console. The output will show each unique NOC and their earliest participation year, sorted as described above.

```python
print(df.groupby('noc')['year'].min().reset_index().sort_values(by=['year', 'noc']))
```

Q2: Find all the songs that were top-ranked (at first position) at least once since the year 2005.



```python
import pandas as pd
import warnings # Hide the warnings
warnings.filterwarnings('ignore')

df=pd.read_csv('https://raw.githubusercontent.com/ipvikas/Interview/main/10283-find-the-top-ranked-songs.csv')

df.head()
```

```python
print(df[(df['year_rank'] == 1) & (df['year'] >= 2005)]['song_name'].unique())
```

This single line of code identifies and prints the unique names of songs that achieved the top rank (position 1) in the charts, starting from the year 2005 onwards. Let's break it down:

1. `df['year_rank'] == 1`: This part filters the DataFrame `df` to select rows where the 'year_rank' column is exactly 1, indicating the song was the top-ranked for that year.
2. `df['year'] >= 2005`: This part further filters the rows to include only those where the 'year' is 2005 or later.
3. `&`: The ampersand operator combines these two conditions, so only rows satisfying *both* criteria (top-ranked AND from 2005 or later) are selected.
4. `df[...]`: This entire conditional expression inside the square brackets filters the original `df` to create a temporary DataFrame containing only the relevant rows.
5. `['song_name']`: From this temporary filtered DataFrame, the 'song_name' column is selected.
6. `.unique()`: This method then extracts all the unique song names from the selected 'song_name' column, effectively removing any duplicate entries.
7. `print(...)`: Finally, this function prints the resulting array of unique top-ranked song names to the console.
